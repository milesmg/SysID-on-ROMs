# Cahn–Hilliard ROM stability

Set `dimension`, parameters, and a named initial condition below, then create a FOM-versus-ROM GIF.


```julia
repo_root = normpath(joinpath(@__DIR__, "..", ".."))
import Pkg
Pkg.activate(joinpath(repo_root, "Julia"))
include(joinpath(repo_root, "src", "Tools", "Visualizations", "visualizations.jl"))
```

```julia
dimension = 1
N = dimension == 1 ? 128 : 64
boundary_condition = "periodic"
parameters = EquationParameters(; ε2=1e-2, sigma=1.0, mean_c=0.0, r=10, m=10)
initial_condition = "default"
tfinal = 2.0
```

```julia
stability = run_rom_stability("ch"; N, tfinal, dimension, boundary_condition, parameters, initial_condition)
gif_path = save_rom_stability_gif(stability; title="Cahn--Hilliard FOM versus ROM", max_frames=45, fps=10, colorbar=true, name="ch_rom_stability.gif")
gif_path
```
