# Reaction–diffusion ROM results

Load one ROM run or a directory of ROM runs; all trajectory data are retained after the first solve.


```julia
repo_root = normpath(joinpath(@__DIR__, "..", "..", ".."))
import Pkg
Pkg.activate(joinpath(repo_root, "Julia"))
include(joinpath(repo_root, "src", "Tools", "Visualizations", "visualizations.jl"))
```

```julia
input_path = joinpath(repo_root, "Data", "example_run")
dir_of_dirs = false
run_dirs = visualization_run_directories(input_path; dir_of_dirs, mode=:rom)
trajectory_data = Dict(run_dir => solve_run_trajectories(run_dir) for run_dir in run_dirs)
selected_run = first(run_dirs)
```

```julia
comparison = plot_trajectory_comparison(trajectory_data[selected_run]; title="Reaction--diffusion ROM trajectory", xlabel="x", ylabel="state", colorbar=true, save_name="rd_rom_comparison.png")
gif_path = save_trajectory_gif(trajectory_data[selected_run]; title="Reaction--diffusion ROM trajectory", max_frames=60, fps=8, colorbar=true, name="rd_rom_comparison.gif")
display(comparison)
gif_path
```

```julia
function_plot = plot_learned_function(selected_run; plot_true=true, title="Reaction--diffusion learned s2", colorbar=true, save_name="rd_rom_learned_function.png")
display(function_plot)
print_final_function_l2_error(selected_run)
```
